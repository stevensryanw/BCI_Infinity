## Method for loading all 52 .mat files
```
import requests
import scipy.io
import numpy as np

#URL of the .mat file
base_url = "https://ftp.cngb.org/pub/gigadb/pub/10.5524/100001_101000/100295/mat_data/s"

#Initializing empty list for holding data of all 52 patients
all_patient_data = []

#Loop through the file numbers from s01 to s52
for file_number in range(1, 53):
    #Creates url depending on the loop iteration
    url = f"{base_url}{file_number:02d}.mat"
    #Fetch the .mat file from the URL
    response = requests.get(url)
    #Check if the request was successful
    if response.status_code == 200:
        #Saving content to local files
        file_name = f"s{file_number:02d}.mat"
        with open(file_name, "wb") as mat_file:
            mat_file.write(response.content)
        #Loading .mat file
        data = scipy.io.loadmat(file_name)
        all_patient_data.append(data)
else:
    print("Failed downloading .mat file")

#Converting the list of patients data into a numpy array
all_patient_data = np.array(all_patient_data)
```
